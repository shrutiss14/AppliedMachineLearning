```python
import pandas as pd
from sklearn.model_selection import train_test_split
```

```python
file_path = "SMSSpamCollection"

df = pd.read_csv(file_path, sep='\t', header=None, names=["Label", "Message"])

# encode labels
df['Label'] = df['Label'].map({'ham': 0, 'spam': 1})
```

```python
df
```

```python
# save raw data
df.to_csv("raw_data.csv", index=False)

print("raw_data.csv saved")
```

```python
# creating train, validation and test splits
df = pd.read_csv("raw_data.csv")

train, test = train_test_split(df, test_size=0.2, random_state=21)
train, val = train_test_split(train, test_size=0.25, random_state=21)

train.to_csv("train.csv", index=False)
val.to_csv("validation.csv", index=False)
test.to_csv("test.csv", index=False)

print("Version 1 split created")
```

```python
# tracking data versions with dvc and git
!dvc add raw_data.csv train.csv validation.csv test.csv

!git add raw_data.csv.dvc train.csv.dvc validation.csv.dvc test.csv.dvc .gitignore

# The first version of the dataset split was committed using Git, ensuring that both the code and the corresponding dataset version 
# are reproducible.
!git commit -m "Version 1 dataset split with seed 21"
```

```python
df = pd.read_csv("raw_data.csv")

# The dataset was split again using a different random seed, creating a new version of the dataset. 
# This demonstrates how DVC tracks changes in data versions.

train, test = train_test_split(df, test_size=0.2, random_state=77)
train, val = train_test_split(train, test_size=0.25, random_state=77)

train.to_csv("train.csv", index=False)
val.to_csv("validation.csv", index=False)
test.to_csv("test.csv", index=False)

print("Version 2 split created")
```

```python
!dvc add train.csv validation.csv test.csv

!git add train.csv.dvc validation.csv.dvc test.csv.dvc

# The updated dataset version was tracked and committed, allowing multiple dataset versions to be stored and retrieved.

!git commit -m "Version 2 dataset split with seed 77"
```

```python
def print_distribution():
    for name in ["train.csv", "validation.csv", "test.csv"]:
        df = pd.read_csv(name)
        counts = df["Label"].value_counts()
        print(f"\n{name}")
        print(f"Number of 0s: {counts.get(0,0)}")
        print(f"Number of 1s: {counts.get(1,0)}")
```

```python
# The distribution of the target variable was printed for train, validation, and test sets. 
# This confirms that different dataset versions produce different splits and validates successful version retrieval.

print("VERSION 1 DISTRIBUTION")
print_distribution()
```
